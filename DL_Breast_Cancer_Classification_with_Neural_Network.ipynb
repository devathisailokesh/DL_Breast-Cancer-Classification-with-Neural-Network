Breast Cancer Classification With a simple Neural Network(NN)

**Importing the Dependencies**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

**Data Collection and Processing**

```python
# loading the data from sklearn
breast_cancer_dataset =  sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# Loading the data to a dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the target column to the dataframe
data_frame['label'] = breast_cancer_dataset.target
```

```python
# print the last 5 rows of the dataframe
data_frame.tail()
```

```python
# number of rows and columns in the dataset
data_frame.shape
```

```python
# Getting some Information about the data
data_frame.info()
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
#Statistical Measures about the data
data_frame.describe()
```

```python
# Checking the distribution of Target Varibale
data_frame['label'].value_counts()
```

1 --> Benign
0 --> Malignant

```python
data_frame.groupby('label').mean()
```

Separting the features and target

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data & Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y , test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.fit_transform(X_test)
```

**Building the Neural Network**

```python
# importing the tensorflow and Keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of Neural Network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),  # Flatten the input to a 1D array(Input Neurons)
    keras.layers.Dense(20, activation='relu'),  # Fully connected layer with 20 neurons and ReLU activation(Hidden Layers)
    keras.layers.Dense(2, activation='sigmoid')  # Fully connected layer with 2 neurons and sigmoid activation for binary classification(Output Neurons)
])
```

```python
# Compiling the Neural Network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the Neural Network
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs = 10)
```

Visualizing accuracy and loss

```python
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])

plt.title('model_accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation_data'], loc='lower right')
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])

plt.title('model_loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation_data'], loc='lower left')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
```
